```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

## Countplot

```python
sns.set_style('whitegrid')
plt.figure(figsize = (12, 8))
sns.countplot(x = 'day', data = tips)
```

```python
sns.countplot(tips['day'])
```
